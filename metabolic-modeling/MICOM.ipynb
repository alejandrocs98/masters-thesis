```python
import pandas as pd
from micom.workflows import build
from micom.qiime_formats import load_qiime_model_db, load_qiime_medium
```

```python
taxonomy_names = {'B_caccae': 'Bacteroides caccae',
                'B_cellulosilyticus_WH2': 'Bacteroides cellulosilyticus',
                'B_ovatus': 'Bacteroides ovatus',
                'B_thetaiotaomicron': 'Bacteroides thetaiotaomicron',
                'B_uniformis': 'Bacteroides uniformis',
                'B_vulgatus': 'Bacteroides vulgatus',
                'C_aerofaciens': 'Collinsella aerofaciens',
                'C_scindens': 'Clostridium scindens',
                'C_spiroforme': 'Clostridium spiroforme',
                'D_longicatena': 'Dorea longicatena',
                'P_distasonis': 'Parabacteroides distasonis',
                'R_obeum': 'Blautia obeum'}
id_names = {'B_caccae': 'Bacteroides_caccae',
                'B_cellulosilyticus_WH2': 'Bacteroides_cellulosilyticus',
                'B_ovatus': 'Bacteroides_ovatus',
                'B_thetaiotaomicron': 'Bacteroides_thetaiotaomicron',
                'B_uniformis': 'Bacteroides_uniformis',
                'B_vulgatus': 'Bacteroides_vulgatus',
                'C_aerofaciens': 'Collinsella_aerofaciens',
                'C_scindens': 'Clostridium_scindens',
                'C_spiroforme': 'Clostridium_spiroforme',
                'D_longicatena': 'Dorea_longicatena',
                'P_distasonis': 'Parabacteroides_distasonis',
                'R_obeum': 'Blautia_obeum'}
```

```python
lf0_reads = pd.read_table('../ecological-modeling/mcnulty-datasets/LF0/reads.tsv', index_col=0, sep='\t')
lf0_reads_34 = lf0_reads.loc[:,['-34' in col for col in lf0_reads.columns]].copy()
lf0_reads_34.reset_index(inplace=True)
lf0_reads_34 = lf0_reads_34.melt(id_vars=['name'], var_name='sample', value_name='reads')
lf0_reads_34['sample'] = lf0_reads_34['sample'].str.replace('-34', '')
lf0_reads_34['species'] = lf0_reads_34['name'].map(taxonomy_names)
lf0_reads_34['name'] = lf0_reads_34['name'].map(id_names)
lf0_reads_34.columns = ['id', 'sample_id', 'abundace', 'species']
lf0_reads_34
```

```python
hf0_reads_12 = pd.read_table('../ecological-modeling/mcnulty-datasets/HF0/reads.tsv', index_col=0, sep='\t')
hf0_reads_12.loc[:,['-12' in col for col in hf0_reads_12.columns]]
hf0_reads_12.reset_index(inplace=True)
hf0_reads_12 = hf0_reads_12.melt(id_vars=['name'], var_name='sample', value_name='reads')
hf0_reads_12['sample'] = hf0_reads_12['sample'].str.replace('-12', '')
hf0_reads_12['species'] = hf0_reads_12['name'].map(taxonomy_names)
hf0_reads_12['name'] = hf0_reads_12['name'].map(id_names)
hf0_reads_12.columns = ['id', 'sample_id', 'abundace', 'species']
hf0_reads_12
```
