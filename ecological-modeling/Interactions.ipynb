```python
inter_bayes = pd.read_table('mcnulty-results/mcnulty-LF0-seed96/posteriors/interactions/bayes_factors.tsv', index_col=0)
inter_bayes_mask = inter_bayes > 3
inter_bayes_mask = inter_bayes_mask.values
inter_bayes_mask
```

```python
interactions = pd.read_table('mcnulty-results/mcnulty-LF0-seed96/posteriors/interactions/mean_matrix.tsv', index_col=0)
interactions_matrix = interactions.values
interactions_matrix
```

```python
interactions_masked = np.ma.masked_array(interactions_matrix, mask=~inter_bayes_mask)
interactions_masked = interactions_masked.filled(np.nan)
interactions_masked = pd.DataFrame(interactions_masked, index=interactions.index, columns=interactions.columns)
```
